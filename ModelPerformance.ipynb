```python
y=df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y = train_test_split(X,y,test_size=.3,random_state=1)
```

```python
#print(train_X.shape)
#print(train_y.head())
test_X
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
lin = LinearRegression()
model= lin.fit(train_X,train_y)
y_pred = model.predict(test_X)
#print(y_pred)
#print(test_y)
#print("accuracy_score",accuracy_score(test_y,y_pred))
mse= (math.sqrt(mean_squared_error(y_pred,test_y)))
print("Mean Squared error", round(mse,2))
```

```python
#Mean absolute errot
MAPE= np.mean(abs((y_pred-test_y)/test_y)*100)
print("Mean absolute error",round(MAPE,2))
```

```python
#R-Squared
from sklearn.metrics import  r2_score
r2_score = r2_score(y_pred,test_y)
print('R2 score:',round(r2_score,3))
```

```python
#Adrjuste R2 score
#1- (1 - r2) * (N-1)/(N-p-1)
N=X.shape[0]

p=X.shape[1]

Adj_r2 = (1-r2_score) * ((N-1)/(N-p-1))
print("Adj r2score ",round(Adj_r2,3))
```
